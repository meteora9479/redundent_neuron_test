```lua
object = torch.load( 'VGG_n01443537.t7' )
```

```lua
Plot = require 'itorch.Plot'
local n = 0
local max_increase = 0
increase_curve = {}
point_index = {}

for top_n=1, 512 do 
    local total_increase = 0
    point_index[top_n] = top_n
    
    for i in pairs(object) do 
        local output_org = object[i].insertion_curve[512] 
        local topN_increase = object[i].insertion_curve[top_n] - output_org
        total_increase = total_increase + topN_increase
    end
    
    --print( total_increase )
    increase_curve[top_n] = total_increase
    
    if top_n == 1 or max_increase < total_increase then
        max_increase = total_increase
        n = top_n
    end
end

print('TopN is '.. n .. '    increase:'..max_increase)
plot = Plot():line( point_index, increase_curve,'red','TopN Feature Map Curve'):legend(true):title('TopN is '..n):draw()
```
